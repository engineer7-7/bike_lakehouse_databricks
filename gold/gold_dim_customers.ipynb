- Join Both customers tables (erp-crm)

```python
df_joined_customers = spark.sql(
  """
  SELECT c1.customer_key,
  c1.first_name,
  c1.last_name,
  c1.marital_status,
  c2.gender,
  c1.create_date,
  c2.birth_date
  FROM silver.crm_cust_info c1
  JOIN silver.erp_customer_az12 c2
  ON c1.customer_key = c2.customer_id
  """
)
```

- Check the length of the joined df

```python
df_joined_customers.count()
```

```python
df_joined_customers.select("customer_key").count() == df_joined_customers.count()
```

```python

df_joined_customers.write.format("delta").mode("overwrite").saveAsTable("gold.dim_customers")
```
