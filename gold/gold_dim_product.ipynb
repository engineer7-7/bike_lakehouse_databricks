```python
%sql
SELECT product_key, category
FROM silver.erp_px_cat_g1v2
GROUP BY product_key, category
```

- Filter the erp product table, keep only the unique product_key and category

```python
df_erp_filtered_product = spark.sql(
    """
    SELECT product_key, category
    FROM silver.erp_px_cat_g1v2
    GROUP BY product_key, category

    """
)
```

```python
df_erp_filtered_product.display()
```

```python
# add the crm  table
df_crm_product = spark.table(
    "silver.crm_prd_info"
)
```

```python
df_crm_product.display()
```

- Join the tables

```python
df_joined_products = df_crm_product.join(
  df_erp_filtered_product, "product_key", "left"
)
```

```python
df_joined_products.display()
```

```python
df_joined_products.columns
```

```python
df_joined_products.select("product_key").count() == df_joined_products.count()
```

- Extract as gold table

```python
df_joined_products.write.format("delta").mode("overwrite").saveAsTable("gold.dim_products")
```
