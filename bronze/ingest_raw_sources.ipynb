# read all 6 csv files

```python
# create a function which reads the csv files

def read_csv_file(filename):
  return spark.read.csv(filename, header=True, inferSchema=True)
```

```python
# read the erp source csv files

df_cust = read_csv_file('/Volumes/workspace/bronze/raw_sources/CUST_AZ12.csv')
df_loc = read_csv_file('/Volumes/workspace/bronze/raw_sources/LOC_A101.csv')
df_px = read_csv_file('/Volumes/workspace/bronze/raw_sources/PX_CAT_G1V2.csv')
```

```python
# read the crm source csv files
df_cust_info = read_csv_file('/Volumes/workspace/bronze/raw_sources/cust_info.csv')
df_prd = read_csv_file('/Volumes/workspace/bronze/raw_sources/prd_info.csv')
df_sales = read_csv_file('/Volumes/workspace/bronze/raw_sources/sales_details.csv')
```

# store all the 6 csv files as tables

```python
# function to store the csv files, as tables

def csv_to_table(dataframe, tablename):
  return (dataframe.write
     .format("delta")
     .mode("overwrite")  
     .saveAsTable(f"bronze.{tablename}")
    )
```

```python
# first store the erp source files
csv_to_table(dataframe=df_cust, tablename='erp_cust_az12')
csv_to_table(dataframe=df_loc, tablename='erp_loc_a101')
csv_to_table(dataframe=df_px, tablename='erp_px_cat_g1v2')

# store the crm source files
csv_to_table(dataframe=df_cust_info, tablename='crm_cust_info')
csv_to_table(dataframe=df_prd, tablename='crm_prd_info')
csv_to_table(dataframe=df_sales, tablename='crm_sales_details')
```
